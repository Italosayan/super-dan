```python
import pandas as pd
from IPython.core.display import display, HTML
import pandas_profiling
```

```python
queried = '/Users/italosayan/Code/super_dan/super_dan_app/dataset/queried_data/2019-11-21|14:08:05.401876_crimes.csv'

queried = pd.read_csv(queried)

queried = queried[queried['Statute_Text'].isin(["Larceny", "Burglary"])]

    # Filter interesting columns
cols = ['OccurredFrom_Timestamp',
            'OccurredThrough_Timestamp',
            'Reported_Timestamp',
            'Location_Type',
            'x', 'y',
            'Statute_Text']

queried = queried[cols]
queried = queried.dropna()
```

Training is happening. Is going to be great

```python
features = queried.loc[:, queried.columns != 'Statute_Text']
target = queried.loc[:, 'Statute_Text']
```

```python
features = features.reset_index(drop=True)
location_types = ["Single Family Home",
"Other Outside Location",
"Multiple Dwelling",
"Street	",
"Parking Lot"]

features.loc[~features['Location_Type'].isin(location_types),'Location_Type'] = "Other"
from sklearn.preprocessing import LabelBinarizer
dummy_locations = pd.get_dummies(features.loc[:,"Location_Type"])
features = pd.concat([features, dummy_locations], axis=1).drop(columns="Location_Type")

features = features.dropna()
```

```python
pandas_profiling.ProfileReport(features)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, random_state=0)
```

```python
dummy = DummyClassifier(strategy='uniform', random_state=1)
dummy.fit(features_train, target_train)
dummy.score(features_test, target_test)
```

```python
classifier = RandomForestClassifier()
classifier.fit(features_train, target_train)
classifier.score(features_test, target_test)
```
